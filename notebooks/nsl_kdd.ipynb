```python
train_df['binary_attack'] = train_df['class'].apply(lambda x: 0 if x == 'normal' else 1)
```

```python
# Distribution of binary class
plt.figure(figsize=(8,6))
sns.countplot(x='binary_attack', data=train_df, palette = 'pastel')
plt.title('Distribution of Network raffic (0: Normal, 1:Attack)', fontsize=16)
plt.xlabel('Traffic type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks([0,1], ['Normal', 'Attack'])
plt.show()

print(train_df['binary_attack'].value_counts(normalize=True))
```
