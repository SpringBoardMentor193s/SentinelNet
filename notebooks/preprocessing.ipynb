In this notebook, I will go through the basic preprocessing steps for the NSL-KDD dataset.  
The main steps are:  
1. Handling missing values  
2. Removing duplicates  
3. Encoding categorical features  
4. Scaling numerical features  

These steps are essential to prepare the dataset for machine learning models.

```python

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load dataset
file_path = "../data/KDDTrain+.txt" 
df = pd.read_csv(file_path, header=None)

# Preview dataset
df.head()
```

## Step 1: Add column names
The dataset does not come with predefined column names.  
We assign them based on the NSL-KDD documentation.

```python
columns = [
    'duration','protocol_type','service','flag','src_bytes','dst_bytes','land','wrong_fragment','urgent','hot',
    'num_failed_logins','logged_in','num_compromised','root_shell','su_attempted','num_root','num_file_creations',
    'num_shells','num_access_files','num_outbound_cmds','is_host_login','is_guest_login','count','srv_count',
    'serror_rate','srv_serror_rate','rerror_rate','srv_rerror_rate','same_srv_rate','diff_srv_rate','srv_diff_host_rate',
    'dst_host_count','dst_host_srv_count','dst_host_same_srv_rate','dst_host_diff_srv_rate',
    'dst_host_same_src_port_rate','dst_host_srv_diff_host_rate','dst_host_serror_rate',
    'dst_host_srv_serror_rate','dst_host_rerror_rate','dst_host_srv_rerror_rate','label','difficulty_level'
]

df.columns = columns
df.head()
```

## Step 2: Handle missing values
We check for missing values and drop or fill them if necessary.

```python
# Check for missing values
print(df.isnull().sum().sum(), "missing values.")

df = df.dropna()
```

## Step 3: Remove duplicate rows
To avoid bias and redundancy, we drop duplicate entries.

```python
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]

print(f"Removed {before - after} duplicate rows.")
```

## Step 4: Encode categorical features
The dataset has categorical columns such as `protocol_type`, `service`, and `flag`.  
We use one-hot encoding to convert them into numerical form.

```python
categorical_cols = ['protocol_type', 'service', 'flag']
df = pd.get_dummies(df, columns=categorical_cols)

df.head()
```

## Step 5: Scale numerical features
Since features have different ranges, we normalize them using MinMaxScaler.  
This brings all values between 0 and 1.

```python
scaler = MinMaxScaler()
numeric_cols = df.drop(columns=['label']).columns

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
df.head()
```

## Final Step: Save the processed dataset
We save the cleaned dataset so it can be reused without repeating preprocessing.

```python
output_path = "../data/KDDTrain+_processed.csv"
df.to_csv(output_path, index=False)
print("Preprocessed data saved to:", output_path)
```
